# Histograma das notas e redação

```python
#Notas (dividindo os dados em 50 partes)
base['nota'].plot.hist(bins=50, edgecolor='black')
```

```python
#Notas da Redação (dividindo os dados em 30 partes)
base['nota_redacao'].plot.hist(bins=30, edgecolor='black')
```

# Analisando alguns dados

```python
#Alunos "Vencedores" (Nota maior ou igual a 5 e Nota da redação maior que 0)
base [(base['nota'] >= 5) & (base['nota_redacao'] > 0)]
```

```python
#Média agrupada da nota pelo local de realização da prova 
base.groupby('prova_realizada_em').mean() ['nota'].sort_values()
```

```python
#Média agrupada da nota da redação pelo local de realização da prova 
base.groupby('prova_realizada_em').mean() ['nota_redacao'].sort_values()
```

```python
#Média agrupada da nota pela escolaridade
base.groupby('escolaridade').mean() ['nota'].sort_values()
```

```python
#Média agrupada da nota da redação pela escolaridade
base.groupby('escolaridade').mean() ['nota_redacao'].sort_values()
```

```python
#Média agrupada da nota pelo estado civil
base.groupby('estado_civil').mean() ['nota'].sort_values()
```

```python
#Média agrupada da nota da redação pelo estado civil
base.groupby('estado_civil').mean() ['nota_redacao'].sort_values()
```

```python
#Média agrupada da nota pelo sexo
base.groupby('sexo').mean() ['nota'].sort_values()
```

```python
#Média agrupada da nota da redação pelo sexo
base.groupby('sexo').mean() ['nota_redacao'].sort_values()
```
